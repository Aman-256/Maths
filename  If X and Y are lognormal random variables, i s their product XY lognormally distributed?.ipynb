```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate 100,000 samples from a standard normal distribution
Z1 = np.random.normal(loc=0, scale=1, size=100000)
Z2 = np.random.normal(loc=0, scale=1, size=100000)

# Step 2: Transform to lognormal variables
X = np.exp(Z1)
Y = np.exp(Z2)

# Step 3: Multiply the lognormal variables
XY = X * Y

# Step 4: Take logs again
log_X = np.log(X)
log_Y = np.log(Y)
log_XY = np.log(XY)

# Step 5: Plot the histograms
plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
sns.histplot(log_X, bins=50, kde=True, color='skyblue')
plt.title('Log(X) ~ Normal')

plt.subplot(1, 3, 2)
sns.histplot(log_Y, bins=50, kde=True, color='lightgreen')
plt.title('Log(Y) ~ Normal')

plt.subplot(1, 3, 3)
sns.histplot(log_XY, bins=50, kde=True, color='salmon')
plt.title('Log(XY) = Log(X) + Log(Y) ~ Normal')

plt.tight_layout()
plt.show()
```

```python
# Step 1: Generate correlated normal variables
mean = [0, 0]
cov = [[1, 0.9], [0.9, 1]]  # High positive correlation between Z1 and Z2
Z_correlated = np.random.multivariate_normal(mean, cov, 100000)

Z1_corr = Z_correlated[:, 0]
Z2_corr = Z_correlated[:, 1]

# Step 2: Transform to lognormal variables
X_corr = np.exp(Z1_corr)
Y_corr = np.exp(Z2_corr)

# Step 3: Multiply the lognormal variables
XY_corr = X_corr * Y_corr

# Step 4: Take logs again
log_X_corr = np.log(X_corr)
log_Y_corr = np.log(Y_corr)
log_XY_corr = np.log(XY_corr)

# Step 5: Plot the histograms
plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
sns.histplot(log_X_corr, bins=50, kde=True, color='skyblue')
plt.title('Log(X) (Correlated)')

plt.subplot(1, 3, 2)
sns.histplot(log_Y_corr, bins=50, kde=True, color='lightgreen')
plt.title('Log(Y) (Correlated)')

plt.subplot(1, 3, 3)
sns.histplot(log_XY_corr, bins=50, kde=True, color='salmon')
plt.title('Log(XY) (Correlated)')

plt.tight_layout()
plt.show()
```
