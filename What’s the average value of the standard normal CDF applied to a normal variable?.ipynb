```python
# Re-import necessary libraries after code execution state reset
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Parameters for X ~ N(mu, sigma^2)
mu = 1.0
sigma = 2.0
n_samples = 100000

# Step 1: Generate samples of X ~ N(mu, sigma^2)
X = np.random.normal(loc=mu, scale=sigma, size=n_samples)

# Step 2: Compute Y = Φ(X) where Φ is the standard normal CDF
Y = norm.cdf(X)

# Step 3: Compute the expected value of Y
E_Y_simulated = np.mean(Y)

# Step 4: Compute the theoretical expectation
E_Y_theoretical = norm.cdf(mu / np.sqrt(1 + sigma**2))

# Step 5: Plot histogram of Y values
plt.figure(figsize=(8, 5))
sns.histplot(Y, bins=50, kde=True, color="cornflowerblue")
plt.title(f"Histogram of Y = Φ(X)\nSimulated E[Y] ≈ {E_Y_simulated:.4f}, Theoretical E[Y] ≈ {E_Y_theoretical:.4f}")
plt.xlabel("Y = Φ(X)")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```
