```python
# Re-import necessary libraries after reset
import numpy as np
import matplotlib.pyplot as plt

# Settings
np.random.seed(42)
n_steps = 1000
n_simulations = 3

# Generate martingale (fair game) paths
martingale_paths = [np.cumsum(np.random.normal(0, 1, n_steps)) for _ in range(n_simulations)]

# Generate trending (non-martingale) paths by adding upward drift
trending_paths = [np.cumsum(np.random.normal(0.1, 1, n_steps)) for _ in range(n_simulations)]

# Plotting
plt.figure(figsize=(14, 6))

# Martingale plot
plt.subplot(1, 2, 1)
for path in martingale_paths:
    plt.plot(path)
plt.title("Martingale (No Trend - Fair Game)")
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.axhline(0, color='black', linestyle='--')

# Trending process plot
plt.subplot(1, 2, 2)
for path in trending_paths:
    plt.plot(path)
plt.title("Non-Martingale (Trending Process)")
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.axhline(0, color='black', linestyle='--')

plt.tight_layout()
plt.show()
```
