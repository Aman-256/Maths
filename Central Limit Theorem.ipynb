```python
# Re-import required libraries after reset
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulation parameters
n_samples = 10000
sample_sizes = [1, 2, 5, 30, 100]  # Increasing sample sizes to observe convergence
exp_scale = 1.0  # Mean of exponential distribution = 1 / lambda

# Set up the figure
plt.figure(figsize=(14, 10))

for i, n in enumerate(sample_sizes, 1):
    # Generate n_samples of size-n means from exponential distribution
    samples = np.random.exponential(scale=exp_scale, size=(n_samples, n))
    sample_means = samples.mean(axis=1)

    # Plot histogram
    plt.subplot(2, 3, i)
    sns.histplot(sample_means, bins=40, kde=True, stat="density", color='cornflowerblue')
    plt.title(f'Sample Size = {n}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.axvline(x=1, color='red', linestyle='--', label='True Mean = 1')
    plt.legend()

plt.tight_layout()
plt.suptitle("Central Limit Theorem with Exponential Distribution", fontsize=16, y=1.02)
plt.show()
```
