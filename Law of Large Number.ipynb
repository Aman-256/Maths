```python
# Re-import necessary libraries after code execution environment reset
import numpy as np
import matplotlib.pyplot as plt

# Simulate coin flips (1 = Head, 0 = Tail)
np.random.seed(42)
n_trials = 10000
coin_flips = np.random.choice([0, 1], size=n_trials)

# Compute running average
running_avg = np.cumsum(coin_flips) / np.arange(1, n_trials + 1)

# True mean of a fair coin
true_mean = 0.5

# Plotting the convergence
plt.figure(figsize=(10, 6))
plt.plot(running_avg, label='Running Average of Heads')
plt.axhline(true_mean, color='red', linestyle='--', label='True Mean = 0.5')
plt.title('Law of Large Numbers: Convergence of Coin Flip Average to True Mean')
plt.xlabel('Number of Flips')
plt.ylabel('Average')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
